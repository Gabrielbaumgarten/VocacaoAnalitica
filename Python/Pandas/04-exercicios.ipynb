### 1. Some todos os números de 1 até 100

```python
sum(range(1,101))
```

### 2. Some todos os números pares de 1 até 100

```python
soma = [x for x in range(1,101) if x % 2 == 0]
sum(soma)
# sum(range(0,101, 2))
```

```python
soma = [x for x in range(1,101) if x % 2 != 0]
sum(soma)
# sum(range(1,101, 2))
```

### 4. Dada a lista abaixo, some todos os elementos inteiros positivos

```python
lista = [-2, 2, 'celular', 10, -99, False]
lista
```

```python
positivos = [x for x  in lista if type(x) == int and x >= 0]
sum(positivos)
```

### 5. Dada a String

```python
string = 'pYthOn É A mELhoR LiNguaGEM De PRoGRaMaçãO'
string
```

Formate-a com as primeiras letras maiúsculas:

```python
frase = ""
for palavra in string.split(" "):
    frase += palavra.capitalize() + " "

frase
```

Com todas as letras minúsculas

```python
string.lower()
```

Com todas as letras maiúsculas

```python
string.upper()
```
