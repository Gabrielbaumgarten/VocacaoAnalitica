### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```python
import numpy as np

array = np.random.randint(50,size=10)
print('Média: ',array.mean())
print('Desvio Padrão: ',round(array.std(), 2))
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```python
array = np.random.rand(10)

print('Original: ',array)

for i in range(len(array)):
    if array[i] < 0.5:
        array[i] = np.sqrt(array[i])

print('Raiz: ',array)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```python
matriz = np.random.randint(50, size=[3,3,3])
print(matriz)

for i in range(3):
    for j in range(3):
        print("Maior valor da linha {} coluna {}: {}".format(i, j, matriz[i,j].max()))
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```python
matriz = np.random.randint(50, size=[5,5])

media = np.mean(matriz, axis=1)

result = matriz - media[:, np.newaxis]

result
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```python
matriz = np.random.randint(50, size=[4,4])

print(matriz)
print('Média diagonal principal: ',matriz.diagonal().mean())
print('Desvio padão diagonal principal: ',matriz.diagonal().std())

print('Média diagonal secundária: ',np.fliplr(matriz).diagonal().mean())
print('Desvio padão diagonal secundária: ',np.fliplr(matriz).diagonal().std())
```
